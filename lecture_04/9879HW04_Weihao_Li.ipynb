(a) Compute the time series of trade signs $\epsilon_t$ according to $\epsilon_t=+1$ if $p_t>m_t$ and  $\epsilon_t=-1$ otherwise, using the Lee-Ready rule for trades at mid.  Name this series *tradeSigns*. (Note this was already done in the lecture.)

(b) Compute the Lee-Ready trade-sign indicator time series (as if you didn't have quotes and so no mid-prices).

(c) Compute the correlation of these two indicators.

(d) Assume the rule in part (a) is 100% accurate.  What percentage of trades are correctly signed by the Lee-Ready test?

(e) Explain the relationship between your answers in (c) and (d).


```r
#(a)Compute the time series of trade signs 
load("tq.rData")
tqdata <- tqBAC
tradeSigns <- getTradeDirection(tqdata) 

#(b)Compute the Lee-Ready trade-sign indicator time series 
prices <- as.numeric(tqdata$PRICE)
price_length <- length(prices)
lr_indicator <- rep(0,length(price_length-1))
for (i in 2: price_length){
    if (prices[i] > prices[i-1])
        lr_indicator[i-1] <- 1
    else if(prices[i] < prices[i-1])
        lr_indicator[i-1] <- -1
    else
        lr_indicator[i-1] <- lr_indicator[i-2]
        }
        
        
# (c) Compute the correlation of these two indicators
            
lr_cor <- cor(lr_indicator, tradeSigns[-1])
cat("The correlation is:", lr_cor, "\n")

#(d) What percentage of trades are correctly signed by the Lee-Ready test?
right_count = 0
for(i in 1:(price_length-1)){
    
    if (lr_indicator[i] == tradeSigns[i+1]){
       
        right_count = right_count +1
        }
}
cat("The percentage that Lee-Ready is correct is : ",right_count/(price_length-1)*100,'%','\n' )
#e Explain the relationship between your answers in (c) and (d)
    print('large correlation means that one variable can explain more about the other, so more percentage of the indicator will be the same')
```

####  (a) solution

```r
####  (b) solution

tradeSigns <- getTradeDirection(tqBAC) 



eps_t = tradeSigns[1:(length(tradeSigns)-1)]
eps_t1 = tradeSigns[2:length(tradeSigns)]

df_tradeSigns = data.frame(eps_t, eps_t1)
# set intercept as 0, since expectation is zero
df_tradeSigns.lm = lm(eps_t ~  ( eps_t1-1), data=df_tradeSigns)
cat("The AR coefficient is:", coefficients(df_tradeSigns.lm))
```

    * $\beta = (1 - 2\,\pi)$
* A small beta indicate that $\pi$ is closed to 0.5,implying that they are independent
