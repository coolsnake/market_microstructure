## Predictability of order flow

### 1. (6 points)

Suppose that the distribution of meta order sizes $Q$ is exponential so that

$$
\Pr(Q>V)= \frac{1}{\lambda}\,e^{-\lambda\,V}
$$

Assuming slices of  constant size $\bar v$, compute the conditional probability that a hidden order has more slices given that $n$ slices have already been observed.  Compare your result with the case where distribution of metaorder size is power-law.

<font color = 'blue'>

** Answer **

The order size that has already been oberved is $Q_{obs} = \bar v~ n$, thus the probability that the order has more slices is
$$
    \Pr~\left(Q\geq \bar v~(n+1)~|~Q\geq \bar v~n\right) = \frac{\Pr~(Q\geq \bar v~(n+1))}{\Pr~(Q\geq \bar v~n)} = e^{-\lambda \bar v}
$$

If the distribution of metaorder size is power-law, i.e.
$$
    \Pr(Q>V) = V^{-\beta}
$$

then the probability that the order has more slices is
$$
    \Pr~\left(Q\geq \bar v~(n+1)~|~Q\geq \bar v~n\right) = \frac{\Pr~(Q\geq \bar v~(n+1))}{\Pr~(Q\geq \bar v~n)} = \left(\frac{n}{n+1}\right)^\beta
$$

## Price manipulation

### 2. (10 points)

Consider the price process

$$
S_t=S_0+\int_0^t\,f(\dot{x}_s)\,G(t-s)\,ds+\int_0^t\,\sigma\,dZ_s
$$

where ${\dot x}_s$ is the rate of trading in dollars at time $s < t$, $f(\dot{x}_s)$ represents the impact of trading at time $s$ and $G(t-s)$ is a decay factor.

(a) With the choices

$$
f(v)=\sqrt{v},\quad G(\tau)=\frac{1}{1+\tau},
$$
    
repeat the computations in Lecture 8  to derive an expression for the expected cost of accumulating shares at the rate $v_1$ and liquidating them at the rate $v_2$.

(b) Verify that with $T=1$, $v_1=1/4$, $v_2=1$, the expected cost of this round-trip strategy is negative, thus demonstrating that price manipulation is possible. (Recall from the lecture slides that price manipulation is always possible if $f(\cdot)$ is nonlinear and the decay kernel $G(\cdot)$  is finite at the origin.)

Hint: You might consider using Mathematica or Wolfram Alpha to do the integrations.

<font color = 'blue'>

** Answer **

(a)

With the choices $f(v)=\sqrt{v}$ and $G(t-s)=\frac{1}{1+t-s}$,
consider a strategy where shares are accumulated at the constant rate $v_1$ and then liquidated again at the constant
rate $v_2$. The cost of this strategy is given by $C_{11}+C_{22}-C_{12}$ with

\begin{eqnarray*}
C_{11}&=&v_1\,\sqrt{v_1}\,\int_0^{\theta\,T}\,dt\,\int_0^t\,\frac{1}{1+t-s}\,ds=v_1\,\sqrt{v_1}\,\left[(1+\theta T)\ln\,(1+\theta T)-\theta T\right]  \\
C_{22}&=&v_2\,\sqrt{v_2}\,\int_{\theta\,T}^T\,dt\,\int_{\theta\,T}^t\,\frac{1}{1+t-s}\,ds=v_2\,\sqrt{v_2}\,\left\{\left[1+(1-\theta)T\right]\ln\left[1+(1-\theta)T\right]-(1-\theta)T\right\} \\
C_{12}&=&v_2\,\sqrt{v_1}\,\int_{\theta\,T}^T\,dt\,\int_0^{\theta\,T}\,\frac{1}{1+t-s}\,ds=v_2\,\sqrt{v_1}\,\left\{(1+T)\ln(1+T)-(1+\theta T)\ln(1+\theta T)-\left[1+(1-\theta)T\right]\ln\left[1+(1-\theta)T\right]\right\}
\end{eqnarray*}

where $\theta$ is such that $v_1\,\theta\,T-v_2\,(T-\theta\,T)=0$ so
$$\theta= \frac{v_2}{v_1+v_2}$$

Thus
$$
\E [\Pi]=v_1\,\sqrt{v_1}\,\left[(1+\theta T)\ln\,(1+\theta T)-\theta T\right]+v_2\,\sqrt{v_2}\,\left\{\left[1+(1-\theta)T\right]\ln\left[1+(1-\theta)T\right]-(1-\theta)T\right\}-v_2\,\sqrt{v_1}\,\left\{(1+T)\ln(1+T)-(1+\theta T)\ln(1+\theta T)-\left[1+(1-\theta)T\right]\ln\left[1+(1-\theta)T\right]\right\}
$$

<font color = 'blue'>

(b)

When $T=1$, $v_1=1/4$, $v_2=1$, then $\theta=4/5$, and

$$
C_{11}= 0.03225 \\
C_{22}= 0.01878 \\
C_{12}= 0.05475 \\
$$

Thus the cost of this strategy is 

$$
\E[\Pi]=C_{11}+C_{22}-C_{12}=-0.0037<0
$$

So the expected cost of this round-trip strategy is negative, and price manipulation is possible. 

## The Obizhaeva-Wang model

### 3. (8 points)

In the Obizhaeva Wang (OW) model, the price follows

$$
S_t=S_0+\eta\,\int_0^t\,{\dot x}_s\,e^{-\rho\,(t-s)}\,ds+\int_0^t\,\sigma\,dZ_s
$$

(a) Derive an expression for the expected price path corresponding to a VWAP execution.

(b) With the duration $T$ of the meta order set to 1, plot the expected price path from time $t=0$ to time $t=2$ for $\rho=0.5,1,$ and $2$.

(c) What is the expected average cost per share $\bar S_T$ of a VWAP order in the OW model?

<font color = 'blue'>

** Answer **

(a)

Assuming we want to short the underlying asset.

For VWAP, ${\dot x}_s=-X/T$, where $X$ is the total trade size and $T$ is the
duration of the order. So

When $0<t\leq T$,
\begin{align*}
\E S_t &= \E S_0-\E\,\left( \eta\,\frac{X}{T}\int_0^t e^{-\rho(t-s)}ds\right) \\
&= S_0-\frac{\eta X}{\rho T}\,\left(1-e^{-\rho t}\right)
\end{align*}

When $t>T$, ${\dot x}_t=0$, so 
\begin{align*}
\E S_t &= \E S_0-\E\,\left( \eta\,\frac{X}{T}\int_0^T e^{-\rho(t-s)}ds\right) \\
&= S_0-\frac{\eta X}{\rho T}\,\left(e^{-\rho (t-T)}-e^{-\rho t}\right)
\end{align*}

<font color = 'blue'>

(b)

```r
T_du <- 1
eta <- 1
X <- 5
rho_1 <- 0.5
rho_2 <- 1
rho_3 <- 2

t <- seq(0,2,0.01)

St_1 <- c(-eta*X/(rho_1*T_du)*(1-exp(-rho_1*t[t<=T_du])),
          -eta*X/(rho_1*T_du)*(exp(-rho_1*(t[t>T_du]-T_du))-exp(-rho_1*t[t>T_du])))
St_2 <- c(-eta*X/(rho_2*T_du)*(1-exp(-rho_2*t[t<=T_du])),
          -eta*X/(rho_2*T_du)*(exp(-rho_2*(t[t>T_du]-T_du))-exp(-rho_2*t[t>T_du])))
St_3 <- c(-eta*X/(rho_3*T_du)*(1-exp(-rho_3*t[t<=T_du])),
          -eta*X/(rho_3*T_du)*(exp(-rho_3*(t[t>T_du]-T_du))-exp(-rho_3*t[t>T_du])))

plot(t,St_1,xlab="t",ylab="expected price", col="red", type="l")
lines(t,St_2,col="green", type="l")
lines(t,St_3,col="blue", type="l")
legend("bottomleft",pch=-1,lty=1,col=c("red","green","blue"),legend=c("rho = 0.5","rho = 1","rho = 2"))
```

(c)

\begin{align*}
\E[C_T(x)] &= \int_0^T \dot x_t dt \int_0^t f(\dot x_s) G(t-s)ds \\
&= \int_0^T \frac{X}{T} dt \int_0^t \eta \frac{X}{T} e^{-\rho(t-s)}ds \\
&=\frac{\eta}{\rho}\left(\frac{X}{T}\right)^2\int_0^T\,\left(1-e^{-\rho t}\right)dt \\
&=\frac{\eta}{\rho}\frac{X^2}{T}+\frac{\eta}{\rho^2}\frac{X^2}{T^2}\left(e^{-\rho T}-1\right)
\end{align*}

The cost per share of executing an order using VWAP is therefore

$$
\hat{C}=\frac{\eta}{\rho}\frac{X}{T}+\frac{\eta}{\rho^2}\frac{X}{T^2}\left(e^{-\rho T}-1\right)
$$
